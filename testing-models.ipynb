```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
full_dataset = pd.read_csv('train.csv')
dataset = full_dataset[full_dataset['sc_w'] != 0].copy()
```

```python
full_dataset
```

```python
full_dataset.drop('price_range',axis = 1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(full_dataset.drop('price_range',axis = 1), full_dataset['price_range'],test_size = 0.3,random_state=100)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(random_state=100)
random_forest.fit(X_train,Y_train)
random_forest.score(X_test,Y_test)
```

Random Forest Classifier with CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [500,600]
criterion = ["gini", "entropy"]
max_features = ["auto", "sqrt", "log2"]
max_depth = [None,2,4,6,8,10,12]
min_samples_split = np.arange(2, 10, step=2)
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

param_grid = {
    "n_estimators": n_estimators,
    "criterion": criterion,
    "max_features": max_features,
    "max_depth": max_depth,
    "min_samples_split": min_samples_split,
    "min_samples_leaf": min_samples_leaf,
    "bootstrap": bootstrap,
}
```

```python
random_forest = RandomForestClassifier(random_state=100)
random_cv = RandomizedSearchCV(
    random_forest, param_grid, n_iter=5, cv=3, n_jobs=-1, random_state = 100
)
```

```python
random_forest_cv = random_cv.fit(X_train, Y_train)
```

```python
random_forest_cv.best_params_
```

```python
random_forest_best_model = RandomForestClassifier(**random_forest_cv.best_params_, random_state = 100)
random_forest_best_model.fit(X_train,Y_train)
random_forest_best_model.score(X_test,Y_test)
```

Support Vector Machines

```python
from sklearn.svm import SVC
```

```python
support_vector_machine = SVC(random_state=100)
support_vector_machine.fit(X_train,Y_train)
support_vector_machine.score(X_test,Y_test)
```

Support Vector Machine with CV

```python
kernel = ['linear','poly','rbf','sigmoid']
degree = [3,5,7]

param_grid = {
    "kernel": kernel,
    "degree": degree
}
```

```python
support_vector_machine = SVC(decision_function_shape='ovo',random_state=100)
random_cv = RandomizedSearchCV(
    support_vector_machine, param_grid, n_iter=5, cv=3, n_jobs=-1, random_state = 100
)

support_vector_machine_cv = random_cv.fit(X_train, Y_train)
support_vector_machine_cv.best_params_
```

```python
support_vector_machine_model = SVC(**support_vector_machine_cv.best_params_, random_state = 100)
support_vector_machine_model.fit(X_train,Y_train)
support_vector_machine_model.score(X_test,Y_test)
```
